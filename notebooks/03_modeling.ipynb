```python
# 1. Importar librerías
import pandas as pd
import numpy as np

# 2. Carga dataset limpio
DATA_PATH = "../data/processed/diabetic_data_clean.csv"
df = pd.read_csv(DATA_PATH)

# 3. Crear variable exposición  # insulina activa + cambio terapéutico
df["exposicion"] = np.where(
    (df["insulin"] != "No") & (df["change"] == "Ch"),
    1, 0
)

# 4. Construir tabla 2x2  # exposición vs reingreso <30 días
tabla = pd.crosstab(df["exposicion"], df["readmitted_30d"])

# 5. Extraer frecuencias  # a,b,c,d clásicos
a = tabla.loc[1,1]  # expuestos + reingreso
b = tabla.loc[1,0]  # expuestos + no reingreso
c = tabla.loc[0,1]  # no expuestos + reingreso
d = tabla.loc[0,0]  # no expuestos + no reingreso

# 6. Calcular Odds Ratio  # fórmula estándar
OR = (a * d) / (b * c)

# 7. Calcular error estándar en log(OR)  # base estadística
SE = np.sqrt((1/a) + (1/b) + (1/c) + (1/d))

# 8. Intervalo de confianza 95%  # transformación logarítmica
log_or = np.log(OR)
IC_inf = np.exp(log_or - 1.96 * SE)
IC_sup = np.exp(log_or + 1.96 * SE)

# 9. Métricas diagnósticas  # rendimiento de la exposición como marcador
sensibilidad = a / (a + c)
especificidad = d / (b + d)
VPP = a / (a + b)
VPN = d / (c + d)

# 10. Mostrar resultado final
print(f"OR = {OR:.3f}")
print(f"IC95% = ({IC_inf:.3f} – {IC_sup:.3f})")
print(f"Sensibilidad = {sensibilidad:.3f}")
print(f"Especificidad = {especificidad:.3f}")
print(f"VPP = {VPP:.3f}")
print(f"VPN = {VPN:.3f}")
```
